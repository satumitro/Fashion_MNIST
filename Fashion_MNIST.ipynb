```python
import tensorflow as tf
print(tf.__version__)
import numpy as np
import matplotlib.pyplot as plt
```

### **2.Loading MNIST DATASET**

```python
mnist=tf.keras.datasets.fashion_mnist
```

```python
(training_images,training_lebels), (test_images, testlabels) =mnist.load_data()
```

## **3.Representing Data**

```python

np.set_printoptions(linewidth=200)
plt.imshow(training_images[0])
print(training_lebels[0])
print(training_images[0])
```

## **4.Normalizing the DATA**

```python
training_images=training_images/255.0
test_images=test_images/255.0
```

## **`5.Creating Neural Network & Training Data `**

```python
model=tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                  tf.keras.layers.Dense(128,activation=tf.nn.relu),
                                  tf.keras.layers.Dense(10,activation=tf.nn.softmax)])
```

```python
model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
history=model.fit(training_images,training_lebels,epochs=10)
```

```python
history.history.keys()
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'],loc='upper left')
plt.show()
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'],loc='upper left')
plt.show()
```

## **6.Evaluting the DATA**

```python
model.evaluate(test_images, testlabels)
```

# **7.Predict the First 10 images, Probability Distribution**

```python
p=model.predict(test_images[:10])
print(p)
```

```python
pred=np.argmax(p,axis=1)
print(pred)
print(testlabels[:10])
```

## **8.Visulizing the Result**

```python
for i in range(10):
  plt.imshow(test_images[i],cmap='binary')
  plt.title('Orginal:{},Predicted{}'.format(testlabels[i],pred[i]))
  plt.axis("Off")
  plt.figure()
```
